```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#o que j√° sabemos?
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            "McDonald's", 'Visa',
            'Facebook', 'Alibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273  ]
```

```python
plt.rcParams["figure.figsize"] = (15, 8)
plt.plot(empresas, bilhoes)
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilh√µes', fontsize=15)
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xticks(rotation=45)
plt.legend(['Bilh√µes de d√≥lares'], loc = 'right', fontsize=15)
plt.show()

```

```python
# Modificando o tipo da linha

plt.rcParams["figure.figsize"] = (15, 8)
plt.plot(empresas, bilhoes, color = 'green', marker = 'o')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilh√µes', fontsize=15)
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xticks(rotation=45)
plt.legend(['Bilh√µes de d√≥lares'], loc = 'right', fontsize=15)
plt.show()




```

```python
#mudando o tipo do gr√°fico e colocando a legenda acima
plt.bar(empresas, bilhoes, color = 'green')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilh√µes', fontsize=15)
plt.title('Empresas mais ricas do mundo', fontsize=20)
plt.xticks(rotation=45)
plt.legend(['Bilh√µes de d√≥lares'], loc = 'right', fontsize=15)

```
